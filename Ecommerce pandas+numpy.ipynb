# Ecommerce


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'ecommerce.csv')
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df.head()
```

```python
# 1- What are the total Order value for each country
order_per_country = df.groupby('Country')['OrderValue'].sum().reset_index()
order_per_country
```

```python
# 2- What each Major Category sold quantity during November 2011
November = df[df['Year-Month'] == '2011-11']
November.groupby('Major Category').sum()['Quantit'].reset_index()
```

```python
# 3- What's Total Order Value for each Minor Item in Kitchen
SKU_level = df.groupby(['Major Category','Minor Category'])['OrderValue'].sum().reset_index()
SKU_level
```

```python
# 4- What Month/s contributes to high demand in orders
df.groupby('Year-Month')['Quantit'].sum()
```

```python
# 5- What Month/s contributes to high OrderValue?
df.groupby('Year-Month')['OrderValue'].sum()
```

```python
# 6- Why April and December suffer from low sales? 
months = ['2011-04', '2011-12']
April_and_December = df[df["Year-Month"].isin(months)]

April_and_December.groupby(['Year-Month','Major Category','Minor Category','Country'])['OrderValue'].sum().reset_index()
```

```python
# 7- Why May and November have high sales? 
months = ['2011-05', '2011-11']
May_and_November = df[df["Year-Month"].isin(months)]

May_and_November.groupby(['Year-Month','Major Category','Minor Category','Country'])['OrderValue'].sum().reset_index()
```

```python
# 8- Most Major Items sold in April and December
April_and_December.groupby('Major Category').sum()
```

```python
# 9- Most Major Items sold in May and November
May_and_November.groupby('Major Category').sum()
```

```python
# 10- What Hour recives high demand for ordering?
df['Hour'] = df['InvoiceTime'].str.split(':').str[0]
df.groupby('Hour')['Quantit'].sum()
```

```python
# 11- What's the Time of the day that contributes to high Order Values?
def time_of_day(x):
    if x in range(6,12):
        return 'Morning'
    elif x in range(12,16):
        return 'Afternoon'
    elif x in range(16,22):
        return 'Evening'
    else:
        return 'Late night'
df.Hour = pd.to_numeric(df.Hour, errors='coerce')
df['OrderTimeOfDay']=df['Hour'].apply(time_of_day)
df.groupby('OrderTimeOfDay')['Quantit'].sum()
```

```python
# 12 - Order VAlue For each Country in each month
pd.pivot_table(df, values='OrderValue', index=['Year-Month'], columns='Country', aggfunc='sum')
```

```python
# 13 - Order VAlue For each Country and Major Category in Minor Category
pd.pivot_table(df, values='OrderValue', index=['Country','Major Category'], columns='Year-Month', aggfunc='sum')
```

```python
# 14 - Order VAlue For each Country and Major Category in Minor Category
pd.pivot_table(df, values='Quantit', index=['Country','Major Category'], columns='Hour', aggfunc='sum')
```

```python
# 15- Correlation between 'Quantit','OrderValue','UnitPrice'.
features=['Quantit','OrderValue','UnitPrice']
df[features].corr()
```
